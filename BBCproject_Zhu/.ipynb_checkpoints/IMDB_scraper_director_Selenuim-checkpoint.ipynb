```python
df_movie = pd.read_csv("df_movie_all.csv", dtype='object')

df_movie
```

```python
driver = webdriver.Chrome()
import time
def get_country_info(row):
    try:
        driver.get("https://www.imdb.com/")
        time.sleep(1)
        text_input = driver.find_element_by_id("navbar-query")
        time.sleep(1)
        text_input.send_keys(row['movie_dir'])
        time.sleep(1)
        info = driver.find_element_by_class_name("suggestionlabel")
        time.sleep(1)
        info.click()
        time.sleep(1)
        country = driver.find_element_by_id('name-born-info').text

        print("This director is from", country)

        return pd.Series({
            'dir_cn': country
        })
    except:
        return pd.Series({})  
```

```python
new_df_movie = df_movie.apply(get_country_info, axis=1).join(df_movie)
new_df_movie = new_df_movie[['critics_name', 'critics_org', 'critics_cn', 'movie_name', 'movie_dir', 'dir_cn', 'movie_year']]
```

```python
new_df_movie.head(20)
```

```python
new_df_movie['dir_country'] = new_df_movie['dir_cn'].str.extract(r",\s(\w+)$")
new_df_movie.head(10)
```
