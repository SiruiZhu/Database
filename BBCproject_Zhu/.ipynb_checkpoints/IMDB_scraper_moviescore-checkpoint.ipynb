```python
import time
driver = webdriver.Chrome()
def get_movie_info(row):
    try:
        driver.get("https://www.imdb.com/")
        time.sleep(1)
        text_input = driver.find_element_by_id("navbar-query")
        time.sleep(1)
        text_input.send_keys(row['movie_name'])
        time.sleep(1)
        info = driver.find_element_by_class_name("suggestionlabel")
        info.click()
        score = driver.find_element_by_class_name("ratingValue").text
        genres = driver.find_elements_by_class_name("subtext")
        print("This movie's score is", score)

        gendres = driver.find_elements_by_class_name("subtext")
        for gendre in gendres:
            info = gendre.find_elements_by_tag_name('a')
            genre = []
            for word in info[:-1]:
                #print(word.text)
                genre.append(word.text)
    
        return pd.Series({
                'IMDb_score':score,
                'genre': genre
        })    
    except:
        return pd.Series({})  
```

```python
df_movie2 = df_movie.head(10).apply(get_movie_info, axis=1).join(df_movie)
df_movie2 = df_movie2[['critics_name', 'critics_org', 'critics_cn', 'movie_name', 'IMDb_score', 'genre', 'movie_dir', 'movie_year']]
df_movie2
```

```python
def get_url(row):
    direcor = row['movie_dir']
    director_name = direcor.replace(" ", "+")
    director_name
    url = 'http://www.imdb.com/find?ref_=nv_sr_fn&q='+director_name+'&s=all'
    result = requests.get(url).content
    soup_doc = BeautifulSoup(result, "html.parser")
    page = soup_doc.find_all(class_= 'result_text')[0].find('a')['href']
    info = requests.get('page')
    info_doc = BeautifulSoup(info,"html.parser" )
    dir_country = info_doc.find(id='name-born-info')
    
    return pd.Series({
        'url': url        
    })
```

```python
import requests
from bs4 import BeautifulSoup
def get_country_info(row):
    try:
        direcor = row['movie_dir']
        director_name = direcor.replace(" ", "+")
        director_name
        result = requests.get('http://www.imdb.com/find?ref_=nv_sr_fn&q='+director_name+'&s=all').content
        soup_doc = BeautifulSoup(result, "html.parser")
        info = soup_doc.find_all(class_= 'result_text')[0].find('a')['href']
        page = 'https://www.imdb.com'+info

        info_page = requests.get(page).content
        info_page
        page_doc = BeautifulSoup(info_page, "html.parser")
        country = page_doc.find(id='name-born-info').find_all('a')[-1].text
        print(country)

        return pd.Series({
            'dir_country': country      
        })
    except:
        return pd.Series({})       
```

```python
new_df = df_movie.apply(get_country_info, axis=1).join(df_movie)
new_df = new_df[['critics_name', 'critics_org', 'critics_cn', 'movie_name','movie_dir', 'dir_country', 'movie_year']]
new_df
```
